## 4. 深さ優先探索（再帰）を用いた訪問完了順の取得

木構造 $T$ を表す隣接リスト `list_adjlist` と起点 `initnode` が引数として与えられたとき、$T$ に対し再帰を用いた深さ優先探索を行った際に、$T$ の各点を訪問完了する順序を表すリスト `list_node` を返す関数 `DFSInPostOrderUsingRecursion` を作成して下さい。以下の点に注意して解答して下さい。

1. 各点は `0` 以上の整数で表されます。

4. 点 `node1` の子供を再帰で呼び出す際には、リスト `list_adjlist[node1]` に格納されている順番で呼び出す様にして下さい。
なお、各値の大きさは以下の通りとします。

```python
### この行のコメントを改変してはいけません %5-4% ### 
#解答用セル
def DFSInPostOrderUsingRecursion(list_adjlist, initnode):
    list_node = []
    # initnodeの子を再帰で呼び出す
    for adjnode in list_adjlist[initnode]:
        # initnodeの子の再帰
        list1 = DFSInPostOrderUsingRecursion(list_adjlist, adjnode)
        #再帰の結果を統合する
        list_node.extend(list1) # list_node += list1
    #訪問した点の名前(initnode)をlist_nodeに格納
    list_node.append(initnode)
    #結果を返す
    return list_node
```

`networkx` で訪問完了順も取得可能です。以下のセルも `True` が出ることを確認して下さい。

```python
import random, networkx as nx
#ランダムに木を作成する
class FIFOClass:
    def __init__(self):
        self.list_data = []
    def enqueue(self, item1):
        self.list_data.append(item1)
    def dequeue(self):
        return self.list_data.pop(0)
def constructRandomTree(nodenum, maxchildnum):
    random.seed(a=0)
    #隣接リスト
    list_tree_adjlist = [0] * nodenum
    #FIFOを使う
    FIFO = FIFOClass()
    #根は0
    nodecnt = 0
    FIFO.enqueue(nodecnt)
    nodecnt += 1
    while(len(FIFO.list_data) > 0):
        #新しく子供を作成する点
        node1 = FIFO.dequeue()
        #点node1の子供数を決定
        childnum = random.randint(0, maxchildnum)
        #キューに何も入ってない場合、必ず子供は1つ以上
        if len(FIFO.list_data) == 0 and childnum == 0:
            childnum = random.randint(1, maxchildnum)
        list_tree_adjlist[node1] = []
        for adjnode in range(childnum):
            list_tree_adjlist[node1].append(nodecnt)
            FIFO.enqueue(nodecnt)
            nodecnt += 1
            #点の数が指定数に到達したら終了
            if nodecnt == nodenum:
                break
        #点の数が指定数に到達したら終了
        if nodecnt == nodenum:
            #子供がない点のリストを空にしておく
            node1 += 1
            while(node1 < nodenum):
                list_tree_adjlist[node1] = []
                node1 += 1
            break
    
    return list_tree_adjlist
def getNXGraph(nx_g1, list_adjlist):
    for node1, list_adj in enumerate(list_adjlist):
        for node2 in list_adj:
            nx_g1.add_edge(node1, node2)
nodenum_test4 = 10000; maxchildnum_test4 = 5
list_adjlist_test4 = constructRandomTree(nodenum_test4, maxchildnum_test4)
nx_tree4 = nx.DiGraph(); getNXGraph(nx_tree4, list_adjlist_test4)
getNXGraph(nx_tree4, list_adjlist_test4)
initnode_test4 = 0
list_node_nx4 = list(nx.dfs_postorder_nodes(nx_tree4, source=initnode_test4))
list_node4 = DFSInPostOrderUsingRecursion(list_adjlist_test4, initnode_test4)
print(list_node_nx4 == list_node4)
```

## 5. トポロジカルソート

閉路を持たない有向グラフ $G$ を表す隣接リスト `list_adjlist` が引数として与えられたとき、$G$ のトポロジカルソートを表すリスト `list_tpsorted` を返す関数 `topologicalsort` を作成して下さい。以下の点に注意して解答して下さい。

1. 各点は `0` 以上の整数（通し番号）で表されます。
2. トポロジカルソートを表すリスト `list_tpsorted` は頂点の名前を格納しており、`list_tpsorted[i]` に格納されている頂点は `list_tpsorted[j]` （ただし、`j<i`）に格納されているあらゆる頂点への枝を持ちません。
 * 例えば、`list_tpsorted = [1, 3, 2, 0]` である場合、頂点 `3` は頂点 `1` への枝を持つかもしれませんが、`2` や `0` への枝は決して持ちません。 
3. $G$ には隣接点を持たない点が含まれる可能性があります。
4. トポロジカルソートを求めるには授業で扱った深さ優先探索を用いる手法を使って下さい。
5. 深さ優先探索の起点は、通し番号の小さい頂点から順に選んで下さい。
6. 前問4と同様、点 `node1` の子供を再帰で呼び出す際には、リスト `list_adjlist[node1]` に格納されている順番で呼び出す様にして下さい。
7. 前問で解答した深さ優先探索（再帰。訪問完了順）を用いて解答しますが、以下のセルで定義されている関数 `DFSInPostOrderUsingNx` を用いても構いません。
8. （ヒント→）<font color="white">前問で扱った深さ優先探索（再帰。訪問完了順）は木構造に対するものであったため、訪問済みの頂点を考慮せず1回探索を行えば探索が完了しました。しかし、この問はグラフに対して探索を実行する必要があり、また、隣接点を持たない点が含まれる可能性がある為、訪問済みの頂点を記録しておく必要があります。</font>

以下のセルの `...` のところを書き換えて解答して下さい。

```python
# 深さ優先探索（訪問完了順）（networkx）
import networkx as nx
def DFSInPostOrderUsingNx(list_adjlist, initnode):
    if list_adjlist[initnode] == []:
        return [initnode]
    nx_graph1 = nx.DiGraph()
    for node1 in range(0, len(list_adjlist)):
        for node2 in list_adjlist[node1]:
            nx_graph1.add_edge(node1, node2)
    return list(nx.dfs_postorder_nodes(nx_graph1, source=initnode))
```

```python
### この行のコメントを改変してはいけません %5-5% ### 
#解答用セル
def topologicalsort(list_adjlist):
    # list_tpsortedの初期化
    list_tpsorted = []
    # 全ての点を調べる
    for i in range(len(list_adjlist)):
        #  もし調べたい点が訪問済みでなかったら
        if i not in list_tpsorted:
            # 深さ優先探索をする
             list1 = DFSInPostOrderUsingRecursion(list_adjlist, i)
             # 深さ優先探索をした結果をlist_tpsortedに追加する
             for j in list1:
                 if j not in list_tpsorted:
                     list_tpsorted.append(j)
    return list_tpsorted
```

## 6. 家系図の探索

第3回の課題で出てきた様な男性優位の家系図は木構造で表すことができました。しかし、性別を問わずある人物とその人物の子供に枝が存在するとみなすと、家系図は有向グラフで表すことが出来ます。そこでその様な家系図が与えられた場合に最も先祖から子孫までの世代の数が多い（つまり、グラフ上で最長の）経路の距離を求めてみることにしましょう。

木構造やグラフ上の最短距離は幅優先探索を用いて求めることが出来るのでした。では、最長距離を求めるにはどうすれば良いのでしょうか。

実はトポロジカルソートの結果を用いることで求めることができます。

以下で紹介する方法は**動的計画法**と呼ばれる手法を用いたアルゴリズムになっています。これは、分割統治法に類似して、与えられた問題を部分的な問題に分割して、それらの部分的な問題を逐次的に解決して最終的に与えられた問題を解決する手法になっています。

この手法は、（トポロジカルソートの意味で）グラフにおいて最も端に存在している点の最長距離を `0` とし、その隣接点の最長距離は`1`、更にその隣接点は `2` とし、各点までの最長距離（経路）を逐次的に求めていく方法になっています。

次のセルを実行してみて下さい。この図のグラフは前問の解答のテストに用いたグラフですが、前問の指示に従ってトポロジカルソートを実行した場合、`[2, 1, 8, 0, 3, 4, 5, 6, 7]` という結果が得られたはずです。

このグラフではトポロジカルソートの実行順によっては、点 `7` 以外に `4` や `5` が最初（実行結果では最も大きいインデックス）に来る可能性がある点になっています。これらの点（からなる集合を $U = \{ $ `4`, `5`, `7` $\}$ とします) は、これらの点 $U$ を隣接点としてもつ点（$U$ の**先行点**と呼びます。グラフが木構造である場合これは親になります。）が存在しない点となっており、先行点を持たない $U$ の中に最長経路上において最も端に来る点が存在します。

次に $U$ のいずれかの点を先行点としてもつ点 $v$ （ `0`, `2`, `6`, `8` のいずれかの点）を考えましょう。$v$ は $U$ に含まれていませんが、経路の端点の点である $U$ の点が隣接しているので、$v$ がもし最長経路を構成しているのならば、最長経路上において2番目に来る点になっているはずです（1番目、すなわち端点は $v$ に隣接している $U$ の点です）。つまり、$v$ までの最長距離は `1` になっています。ここで $U$ の点は端点になっていますので、その最長距離を `0` とします。

これを順に繰り返して最長距離を求めますが、問題は端点から離れると、迂回路（う回路）の様な経路が現れることです。そこで次の様に考えます。

ここで、点 $w$ の最長距離を求めようとするときに、$w$ に先行する点までの最長距離が全て求まっているとしましょう（これらの点集合を $V$ とします）。その場合、$w$ の最長距離は $V$ に含まれる点の最長距離の値に $+1$ して得られる値の中で最も大きい値になっています。つまり、数式で表すと以下の通りです。

$w$ の最長距離 $ = \max_{ v' \in V } \{ v' $ の最長距離 $ + 1 \}$  

「得られる値の中で最も大きい値」とするのは、$w$ までの経路には複数の経路があり、$w$ の先行点の中には非常に大きい最長距離の点（迂回路を経由する様な点）や、小さい距離の点が存在する可能性がありますので、その為です。

整理しますと、次の様な手順によって有向グラフ $G$ の最長距離を求めることができます。

1. $G$ のトポロジカルソートを求める。
2. 得られたトポロジカルソートの順で各点 $w$ と $w$ の先行点の集合 $V$ に対して、
$w$ の最長距離 $ = \max_{ v' \in V } \{ v' $ の最長距離 $ + 1 \}$  
を求める。ただし、$V$ に含まれる点が存在しない場合は $w$ の最長距離は 0 とする。
3. 各点の最長距離の中で最も大きい値が、そのグラフの最長経路の距離になっている。

トポロジカルソートを求めるのには前問の関数を用いますが、下記のセルの関数 `topologicalsortNx` を用いても構いません。この関数は前問の仕様を満たす返り値を返します。なお、下記の関数の8, 9行目は削除しても構いません。


```python
# トポロジカルソート (networkx)
import networkx as nx
def topologicalsortNx(list_adjlist):
    nx_graph1 = nx.DiGraph()
    for node1, list_adj in enumerate(list_adjlist):
        for node2 in list_adj:
            nx_graph1.add_edge(node1, node2)
    list_res = nx.topological_sort(nx_graph1) # 削除しても良い
    list_res = list(list_res) # 削除しても良い
    list_res.reverse()
    return list_res
```


では、閉路を持たない有向グラフ $G$ を表す隣接リスト `list_adjlist` が引数として与えられたとき、$G$ の各点までの最長距離を格納したリスト `list_D` を返す関数 `getLongestDistancesUsingRecursion` を作成して下さい。以下の点に注意して解答して下さい。

1. 各点は `0` 以上の整数（通し番号）で表されます。
2. $G$ には隣接点を持たない点が含まれる可能性があります。
3. `list_D` の `i` 番目の要素 `list_D[i]` は、点 `i` の最長距離を格納します。


なお、各値の大きさは以下の通りとします。
* $m =$ $G$ の枝の数
* $n =$ $G$ の点の数
* $k =$ $G$ の点の中で最大の隣接点数

以下のセルの `...` のところを書き換えて解答して下さい。

```python
### この行のコメントを改変してはいけません %5-6% ### 
#解答用セル
def getLongestDistancesUsingRecursion(list_adjlist):
    # list_adjlistのトポロジカルソートを求める
    sorted_list = topologicalsort(list_adjlist)
    # topologicalsortの返り値を逆順にする
    sorted_list.reverse()
    # list_Dの初期化
    list_D = [0]*len(sorted_list)
    # トポロジカルソート順に調べる
    for node in sorted_list:
        # wの最長距離の初期化
        max = 0
        # 全ての点を調べる
        for i in range(len(list_adjlist)):
            # nodeに先行点が存在していた場合，wの最長距離を求める
            if node in list_adjlist[i]:
                if list_D[i] + 1 > max:
                    max = list_D[i] + 1
                list_D[node] = max
    return list_D
```

上のセルで解答を作成した後、以下のセルを実行し、実行結果が全て `True` になることを確認して下さい。

```python
list_adjlist_test1=[[1, 2], [], [3], [1]]
print(getLongestDistancesUsingRecursion(list_adjlist_test1)==[0, 3, 1, 2])
list_adjlist_test2=[[1, 2], [2], [], [2]]
print(getLongestDistancesUsingRecursion(list_adjlist_test2)==[0, 1, 2, 0])
list_adjlist_test3=[[2, 8], [], [], [], [2], [0,2], [1], [6, 8], [1]]
print(getLongestDistancesUsingRecursion(list_adjlist_test3)==[1, 3, 2, 0, 0, 0, 1, 0, 2])
```
